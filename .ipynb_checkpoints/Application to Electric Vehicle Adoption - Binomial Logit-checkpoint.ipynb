### Variables
##### Chosen

```python
sns.set_style('darkgrid')
fig = plt.figure(figsize=(8, 5))
ax = sns.countplot(x='CHOSEN', data=data, palette='rainbow', )
ax.set_xticklabels(['Adopted', 'Not Adopted'], fontdict=dict(fontweight='bold', color='dimgrey'))
ax.set_yticklabels([])
ax.set_xlabel(None)
ax.set_ylabel(None)
ax.set_ylim(0, 2600)
ax.set_title('Number of Households that Adopted and Not Adopted\nthe Electirc Vehicle', fontweight='bold',
             color='dimgray')
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + 0.329, p.get_height() + 50),
               color='teal', fontweight='bold')
plt.savefig('figures/household_adopted.png');
```

```python
# Create a dummy variable for CHOSEN
data['CHOSEN_BINARY'] = data['CHOSEN'].apply(lambda x: int(x == 1))
perc = [data['CHOSEN_BINARY'].mean(), 1 - data['CHOSEN_BINARY'].mean()]
explode = (0.1, 0.0)
plt.pie(perc, labels=['Adopted ', 'Not Adopted '], colors=['teal', 'coral'],
        explode=explode, autopct='%1.1f%%', textprops=dict(fontweight='bold'))
plt.title('Percentage of Adoption of the Electric Car')
plt.savefig('figures/chosen-pie.png');
```

The dataset has a clear bias to no adopt the electric vehicle.
